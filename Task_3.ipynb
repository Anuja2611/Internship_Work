# Categorical Variable Visualisation

```python
# Employee count per Department

sns.countplot(data=ds , x="Department" , palette="Set2")
plt.title("Employee count per Department")
```

# interpretation : 
Finanace is more than marketing and engineering

```python
#Gender Distribution
sns.countplot(data=ds , x="Gender" , palette = "coolwarm")
plt.title("Gender Distribution")
```

# Interpretation
Female count is greater than male count

```python
#employee LOcation Distribution

sns.countplot(data=ds , x="Location" , palette="Spectral")
plt.title("Number of Employee by Location")
plt.xticks(rotation=45)
```

# Interpretation : 
Employees are spread across different location, with some locations having more employees than others

```python
#Age Distribution

sns.histplot(ds["Age"] , bins=10 , kde=True)
plt.title("Age Ditrubution")
```

# Interpretation
peak at 25-30 & 45-50 

```python
# Years of Experience Distribution

sns.histplot(ds["Years_of_Experience"], bins=15,kde=True)
plt.title("Years of Experience")
```

# Interpretation
The distribution of years is bimodal , with peak around 4-5 and 10-15 years,indicating two dominant experience groups

```python
# monthly Salary Distribution
sns.boxplot(data=ds , y="Monthly_Salary" , color="lightblue")
plt.title("Salary Spread")
```

# Interpretation 
data is fairly spread,showing salary variation

```python
# Average Salary by Department
sns.barplot(data=ds , x="Department" , y="Monthly_Salary" , estimator=np.mean , palette="pink")
plt.title("Average Salary by Department")
```

# Interpretation
some dapartments have higher performance scores than others which could be due to the type of work or team efficiency

```python
# performane score by Department

sns.boxplot(data=ds , x="Department" , y="Performance_Score" ,palette="pink")
plt.title("Performance Score by Department")
```

# Intrepretation
some departments have higher performance than others , which could be due to the type of work or team efficiency

```python
# Salary Distribution by Gender
sns.violinplot(data=ds , x="Gender" , y="Monthly_Salary" , palette="muted")
plt.title("Salary by Gender")
```

# Interpretation
There is small diffrence in average salary between males and females which may be due to job roles  or experience.

```python

sns.heatmap(ds[["Age" , "Years_of_Experience" , "Monthly_Salary","Performance_Score"]].corr(),annot=True , cmap="coolwarm")
plt.title("Correlation Matrix")
```

# Interpretation
Dark blue indicates negative correlation and others indicates positive correlation

```python
# salary vs Experience Scatter Plot
sns.scatterplot(data=ds , x="Years_of_Experience" , y="Monthly_Salary",hue="Department")
plt.title("Salary vs. Experience by Department")
```

# Interpretation
as years of experience increases salary increases
