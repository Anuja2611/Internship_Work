```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values = np.nan , strategy="mean")
ds[['Age' , 'Salary']]=si.fit_transform(ds[['Age' , 'Salary']])

#The `fit_transform()` function learns the parameters from the data and immediately applies the transformation in a single step.
```

```python
ds
```

```python
ds.columns
```

```python
#Encoding categorical column     (OneHotEndocing)

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)       #search what is sparse_output
encoded_data = ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# Creatign columns

column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data , columns = column_names)
encoded_df
```

```python
# joining original dataset and new dataframe

ds = pd.concat([ds,encoded_df], axis = 1)
ds
```

```python
# Drop actual categorical country column

ds = ds.drop(columns = ['Country'])
ds
```

```python
# labelEncoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age' , 'Salary']] = sc.fit_transform(ds[['Age','Salary']])
```

```python
#Divide dataset into Independent(input) and Dependent(output)

x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```
